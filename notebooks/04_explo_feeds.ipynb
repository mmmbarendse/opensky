```python
import gzip
import json

def load_jsonl(input_path) -> list:
    """
    Read list of objects from a JSON lines file.
    """
    data = []
    with gzip.open(input_path, 'r') as f:
        for line in f:
            try:
                line = f.readline()
                line = line.decode('utf-8')
                line = line.rstrip()
                data.append(json.loads(line))
            except json.JSONDecodeError as e:
                print(f'Error in line: {line}')
                print(e)

    print(f'Loaded {len(data)} lines from {input_path}') 

    cols = ['post_id', 'user_id', 'instance', 'date', 'text', 'langs', 'repost_count', 'reply_to', 'replied_author', 'thread_root', 'thread_root_author', 'quotes', 'quoted_author', 'labels']
    return pd.DataFrame(data)

df = load_jsonl(path_feed_file)
df
```
