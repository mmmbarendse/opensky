```python
# install and import necessary packages 
!pip install pandas wordcloud matplotlib

import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
```

```python
# examine first peak and select most relevant words
first_peak_df = pd.read_csv("replies_first_peak.csv")
first_label = "negative" 
filtered_first_df = first_peak_df[first_peak_df["sent_label"] == first_label]
replies_first = filtered_first_df["text"]
replies_first_joined = " ".join(replies_first.astype(str))
words_first = re.findall(r'\b\w+\b', replies_first_joined.lower())
word_counted_first = Counter(words_first)
most_common_words_first = word_counted_first.most_common(50)
print(most_common_words_first)
relevant_words_first = " ".join([word for word, _ in most_common_words_first])
wordcloud_first = WordCloud(width = 800, height = 200, background_color = "white").generate(relevant_words_first) 

# show word cloud
plt.figure(figsize = (10, 5))
plt.imshow(wordcloud_first, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
